```python
from scipy.spatial.distance import euclidean
#evklid əvəzinə minkovski istifadə edərək nəticəyə baxa bilərsiniz
#from scipy.spatial.distance import minkowski
```

```python
def FindNearestP(new_data, X):
    Dmin = euclidean(new_data, X[0])
    minID = 0
    for i in range(1, X.shape[0]):
        D = euclidean(new_data, X[i])
        if D < Dmin:
            Dmin = D
            minID = i
    return minID


minID = FindNearestP(new_data,X)
```

Yuxarıdakı nümunədə evklid alqoritması vasitəsilə setə daxil olan yeni elementə ən yaxın bir elementi seçdik

```python
def PlotPoints(X,y,new_data=None,new_prediction=None, nearest_points=None):
    plt.figure(figsize=(8, 8))
    plt.scatter(X[:, 0], X[:, 1], c=color, s=80,alpha=0.4)
    if new_data is not None:
        plt.scatter(new_data[:,0],new_data[:,1],c='black', s=100)
    if nearest_points is not None:
        plt.scatter(nearest_points[:,0],nearest_points[:,1],c='blue', s=100,edgecolor='black')
    plt.title("Classification 'KNN'")
    plt.xlabel("x1")
    plt.ylabel("x2")
```

```python
PlotPoints(X,y,new_data=new_data,nearest_points=X[[minID]])
```

```python
def PlotPoints(X,y,new_data=None,new_prediction=None, nearest_points=None):
    plt.figure(figsize=(8, 8))
    plt.scatter(X[:, 0], X[:, 1], c=color, s=80,alpha=0.4)
    if new_data is not None:
        plt.scatter(new_data[:,0],new_data[:,1],c='black', s=100)
    if nearest_points is not None:
        plt.scatter(nearest_points[:,0],nearest_points[:,1],c='blue', s=100,edgecolor='black')
    if new_prediction is not None:
        plt.scatter(new_data[:, 0], new_data[:, 1], c='yellow', s=100, edgecolor='black')
    plt.title("Classification 'KNN'")
    plt.xlabel("x1")
    plt.ylabel("x2")
```

```python
def PredictClass(new_data, X, y):
    minID = FindNearestP(new_data, X)
    return y[minID]
new_data = np.array([[9,3]]) #test sample
minID = FindNearestP(new_data, X)
y_new = PredictClass(new_data, X, y)
```

```python
PlotPoints(X, y, new_data=new_data, new_prediction=[y_new], nearest_points=X[[minID]])
```

```python
from scipy.spatial.distance import cdist
from collections import Counter
def PredictClass(new_data, X, y, k=1):

    dist = cdist(new_data, X)
    N = y[np.argpartition(dist, k, axis=1)[:, :k]]
    return np.array([Counter(row).most_common(1)[0][0] for row in N])
```

```python
y_pred = PredictClass(new_data, X, y, k=3)
```

```python
PlotPoints(X,y,new_data=new_data,new_prediction=y_pred)
```

Anomalizm

İki sinif arasında sərhəd çəkilsə də datasetin anomal olduğu aşkar görünür.100 nümunə dairə seçdiyimiz üçün dataset təlim(training) üçün yetərsizdir.Bir çox hallarda anomal datasetlərlə rastlaşacaqsınız.Bu zaman normallaşdırma və standartlaşdırma metodlarından istifadə etmək məcburiyyətində qalacaqsınız.Gəlin real nümunələrlə bu metodları nəzərdən keçirək

```python
X, y = datasets.make_blobs(n_samples=100, random_state=4, centers=2, cluster_std=2)
X[:, 1] *= 10000
print(X[:10],y[:10])
```

```python
def PlotPScaled(X, y, new_data=None, new_prediction=None, nearest_points=None):
    plt.scatter(X[:, 0], X[:, 1], c=color, s=100, edgecolor='black', alpha=0.4)
    plt.figure(1, figsize=(12, 8))
    plt.title("Classification KNN")
    plt.xlabel("x1")
    plt.ylabel("x2")
    plt.gca().set_aspect('equal', adjustable='datalim')
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
def Result(prediction_y, true_y):
    accuracy = accuracy_score(prediction_y, true_y)
    precision = precision_score(prediction_y, true_y)
    recall = recall_score(prediction_y, true_y)
    f1 = f1_score(prediction_y, true_y)
    print("Accuracy:  {:.3f}\nPrecision: {:.3f}\nRecall:    {:.3f}\nF1-score:  {:.3f}".format(accuracy, precision, recall, f1))
```

```python
PlotPScaled(X,y)
```
