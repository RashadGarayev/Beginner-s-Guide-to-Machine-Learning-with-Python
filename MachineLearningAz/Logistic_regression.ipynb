1) Low Precision/High Recall: (Aşağı Həssaslıq / Yüksək Xatırlatma)

Yalnış pozitivlərin sayını mütləq azaltmadan saxta neqativlərin sayını azaltmaq istədiyimiz tətbiqlərdə, həssaslığın və ya Xatırlatmanın aşağı dəyəri olan qərar dəyərini seçirik. Bunun səbəbi, xərçəngin olmaması sonrakı tibbi xəstəliklər tərəfindən aşkar edilə bilər, ancaq xəstəliyin mövcudluğu rədd edilmiş bir namizəddə aşkar edilə bilməz.

2. High Precision/Low Recall: (Yüksək Həssaslıq/aşağı xatırlatma)

Mütləq saxta neqativlərin sayını azaltmadan saxta pozitivlərin sayını azaltmaq istədiyimiz tətbiqlərdə, həssaslıq və ya geri çağırmanın yüksək dəyəri olan qərar dəyərini seçirik. Məsələn, müştərilərin fərdi reklamlara müsbət və ya mənfi reaksiya verəcəklərini təsnif etsək, müştərinin reklamentinə müsbət reaksiya göstərəcəyinə tam əmin olmaq istəyirik, çünki əks halda mənfi reaksiya müştəridən potensial satışa səbəb ola bilər .

### Sigmoid Funksiyası (Sigmoid activation)

```python
from IPython.display import Image
Image(filename='image/sigmoid-function-300x138.png')
```

Proqnozlaşdırılan dəyərləri ehtimallara uyğunlaşdırmaq üçün sigmoid funksiyasından istifadə edirik. Funksiya hər hansı bir həqiqi dəyəri 0 ilə 1 arasındakı çıxış dəyərini qaytarır.

$$S_{z} = \dfrac{1}{1 + e^{-z}}$$

<strong>
<p>1) s (z)   :  0 ilə 1 arasındakı çıxış (ehtimal qiymətləndirmə)
<p>2) z       : funksiyaya giriş (alqoritminizin proqnozu, məsələn, mx + b)
<p>3) e          : təbii girişin əsası

Logistik reqressiyanı öyrəndiyimiz ərəfədə bir neçə mövzuları da öyrənmiş olacaqsınız

### Gradient descent (Gradient enmə)

Xərcimizi minimuma endirmək üçün Xətti Reqressiyada olduğu kimi gradient enişindən istifadə edirik. BFGS kimi birləşdirici gradient kimi digər daha inkişaf etmiş optimallaşdırma alqoritmləri var

\begin{align}
s'(z) & = s(z)(1 - s(z))
\end{align}

Eyni dərəcədə  və rahat xərc funksiyası törəməsinə gətirib çıxarır:

$$C' = x(s(z) - y)$$

<p>C′     - çəkilərə görə dəyəri törəməsidir
<p>y      - faktiki sinif etiketidir (0 və ya 1)
<p>s (z)  - modelinizin proqnozu
<p>x      - vektorunuzdur

Qradient enmə optimallaşdırma alqaritmidir,loss funksiyasından danışdıqda əslində qradient enmə alqoritmlərinə toxunurduq.ən optimal funksiyalarından sigmoid relu lrelu və s misal çəkə bilərik.

$$MSE=\frac{1}{n}\sum_{i=1}^{n}\left ( \hat{y} - y \right )^2$$

$$\frac{\partial MSE}{\partial \theta}=\frac{1}{n}\sum_{i=1} (y_{i}-\theta^tx_{i})^2 =$$

$$= \frac{2}{n}\sum_{i=1} (y_{i}-\theta^tx_{i})x_{i}    $$

Optimal grad dəyər teta ilə birbaşa əlaqəlidir.teta-nın 0-a doğru yaxınlaşması,MSE nin optimal dəyərini göstərəcəkdir.Və bu da bizə gradient optimal dəyəri verəcək

## Types of logistic regression-Kategoriyaların sayına görə Logistik reqressiya növləri:

1) <strong>binomial</strong>: hədəf dəyişən yalnız iki mümkün növə sahib ola bilər: "qalib" və "zərər", "keçmək" və "uğursuz", "ölü" və "diri" və s. təmsil edə bilən "0" və ya "1" (binar klassifikasiya)

3) <strong>ordinal</strong>-nizamlı: sifariş edilmiş kateqoriyalarla hədəf dəyişənləri ilə birbaşa əlaqəli olması. Məsələn, bir test balı aşağıdakı kimi təsnif edilə bilər: "çox zəif", "yoxsul", "yaxşı", "çox yaxşı". Burada hər kateqoriyaya 0, 1, 2, 3 kimi bir xal verilə bilər.

Qərar necə qəbul edilir!

$$\hat{y} = sign(x\theta)$$

```python
w = np.array([7, -1, 1]) # Ax + By + C = 0
     
# Qeyd: nöqtədən məsafə (x0, y0) -> Ax + By + C  abs(Ax0 + By0 + C) / sqrt(A^2 + B^2)
    
def predict(x, w):
    return np.sign(x.dot(w))
x_pred = np.hstack([np.random.rand(10, 2) * 4 - 2, np.ones(10).reshape(-1, 1)])
y_pred = predict(x_pred, w)

print(x_pred.shape)
print(y_pred.shape)
```

```python
x_naive = np.arange(-0.5, 0.5, 0.1)
# naive boundary
y_naive = 7 * x_naive + 1

plt.scatter(x_pred[:, 0], x_pred[:, 1], c=['orange' if y_i > 0 else 'blue' for y_i in y_pred], marker='x', s=100)
plt.title("Classification")
plt.xlabel("x1 (feature)")
plt.ylabel("x2 (feature)")
plt.scatter(X[:, 0], X[:, 1], c=colors)
plt.plot(x_naive, y_naive, c='red', label='descision boundary naive')
plt.legend();
```

Nəticənin qiymətləndirilməsi

margin - sərhəddə yaxın elementləri dəyərləndirmək bir o qədər çətin olacaq.Elementlər sərhəddən nəqədər uzaq olarsa bir oqədər iki sinif üçün elementləri ayırd edə biləcəyik.Margin,elementin sərhəddə olan məsafəsidir.

$$M_i = y_i \cdot x_i\theta  $$ ( 𝑦=1 və ya 𝑦=−1)

$$M_i \leq 0 \Rightarrow y_i \neq \hat{y_i}$$
$$M_i > 0 \Rightarrow y_i = \hat{y_i}$$

Funksiyanın zərəri(zero-one-loss)

$$f(x) = \begin{cases}
      1, & \text{əgər}\space \hat{y} \neq y, \\
      0, & \text{əgər}\space \hat{y} = y
    \end{cases}$$

Empirik risk:

$$Q(\theta, x) = \frac{1}{n} \sum_{i=1}^{n} f(x) = \frac{1}{n} \sum_{i=1}^{n}[M_i < 0]$$

Problemi yenidən tənzimləyirik

Bir sinif yerinə, bir sinfə aid olma ehtimalını təxmin edirik

$$\hat{p} = \sigma(x\theta)$$

$$\sigma(t)=\frac{1}{1 + exp(-t)}$$

$\sigma - Sigma$

P(Pozitive) sinfinə aid olma ehtimalı olsun

$$OR = \frac{P} {1 - P}$$

$$log {OR} \in R$$

Xətti reqressiya istifadə edərək 𝑙𝑜𝑔𝑂𝑅 üçün bir proqnoz verməyi öyrəndi

$$log OR = x \theta$$

$$P(y_i = 1 | x, \theta) = \frac {OR} {1 + OR} = \frac{e^{x\theta}} {1 + e^{x\theta}} = \sigma(x\theta)$$

$$P(y_i = -1 | x, \theta) = ... = \sigma(-x\theta)$$

$$P(y | x, \theta) = \prod_i {P(y = y_i | x_i, \theta)}$$

```python
t = np.linspace(-10, 10, 100)
sigmoid = 1 / (1 + np.exp(-t))
plt.figure(figsize=(9, 3))
plt.plot([-10, 10], [0, 0], "k-")
plt.plot([-10, 10], [0.5, 0.5], "k:")
plt.plot([-10, 10], [1, 1], "k:")
plt.plot([0, 0], [-1.1, 1.1], "k-")
plt.plot(t, sigmoid, "b-", linewidth=2, label=r"$\sigma(t) = \frac{1}{1 + e^{-t}}$")
plt.xlabel("t")
plt.legend(loc="upper left", fontsize=20)
plt.axis([-10, 10, -0.1, 1.1])
plt.show()
```

```python
xmin, xmax = -4, 4
xx = np.linspace(xmin, xmax, 100)
lw = 2
plt.plot([xmin, 0, 0, xmax], [1, 1, 0, 0], lw=lw, label="Zero-one loss")
plt.plot(xx, np.log2(1 + np.exp(-xx)), color='green', lw=lw, label="Log loss")
plt.xlabel(r"Decision function $f(x)$")
plt.legend(loc="upper right")
plt.ylabel("$L(y=1, f(x))$")
plt.ylim((-1, 6));
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression(solver='lbfgs')
log_reg.fit(X, y)
x0, x1 = np.meshgrid(
        np.linspace(-2, 2, 500).reshape(-1, 1),
        np.linspace(-3, 3, 200).reshape(-1, 1),
    )
X_new = np.c_[x0.ravel(), x1.ravel()]
y_proba = log_reg.predict_proba(X_new)
plt.figure(figsize=(10, 4))
plt.plot(X[y==0, 0], X[y==0, 1], "bs")
plt.plot(X[y==1, 0], X[y==1, 1], "g^")
zz = y_proba[:, 1].reshape(x0.shape)
contour = plt.contour(x0, x1, zz, cmap=plt.cm.brg)
left_right = np.array([-3, 3])
boundary = -(log_reg.coef_[0][0] * left_right + log_reg.intercept_[0]) / log_reg.coef_[0][1]
plt.clabel(contour, inline=1, fontsize=12)
plt.plot(left_right, boundary, "k--", linewidth=3)
plt.xlabel("x1 (feature)")
plt.ylabel("x2 (feature)")
plt.axis([-3, 3, -2, 2])
plt.show()
```
