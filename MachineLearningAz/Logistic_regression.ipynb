2. High Precision/Low Recall: (Yüksək Həssaslıq/aşağı xatırlatma)

Mütləq saxta neqativlərin sayını azaltmadan saxta pozitivlərin sayını azaltmaq istədiyimiz tətbiqlərdə, həssaslıq və ya geri çağırmanın yüksək dəyəri olan qərar dəyərini seçirik. ### Sigmoid Funksiyası (Sigmoid activation)

```python
from IPython.display import Image
Image(filename='image/sigmoid-function-300x138.png')
```

Proqnozlaşdırılan dəyərləri ehtimallara uyğunlaşdırmaq üçün sigmoid funksiyasından istifadə edirik. Funksiya hər hansı bir həqiqi dəyəri 0 ilə 1 arasındakı çıxış dəyərini qaytarır.

$$S_{z} = \dfrac{1}{1 + e^{-z}}$$

<strong>
<p>1) s (z)   :  0 ilə 1 arasındakı çıxış (ehtimal qiymətləndirmə)
<p>2) z       : funksiyaya giriş (alqoritminizin proqnozu, məsələn, mx + b)
<p>3) e          : təbii girişin əsası

Logistik reqressiyanı öyrəndiyimiz ərəfədə bir neçə mövzuları da öyrənmiş olacaqsınız

Qərar sərhədi

Tutaq ki, bu qərar sərhədi boyunca sıfıra bərabər olan bir xətt təyin etdik. Bu yolla, xəttin bir tərəfindəki nöqtələrin hamısı müsbət dəyərləri, digər tərəfdəki nöqtələrin hamısı mənfi dəyərləri götürür.

### Gradient descent (Gradient enmə)

Xərcimizi minimuma endirmək üçün Xətti Reqressiyada olduğu kimi gradient enişindən istifadə edirik. BFGS kimi birləşdirici gradient kimi digər daha inkişaf etmiş optimallaşdırma alqoritmləri var

\begin{align}
s'(z) & = s(z)(1 - s(z))
\end{align}

Eyni dərəcədə  və rahat xərc funksiyası törəməsinə gətirib çıxarır:

$$C' = x(s(z) - y)$$

<p>C′     - çəkilərə görə dəyəri törəməsidir
<p>y      - faktiki sinif etiketidir (0 və ya 1)
<p>s (z)  - modelinizin proqnozu
<p>x      - vektorunuzdur

Qradient enmə optimallaşdırma alqaritmidir,loss funksiyasından danışdıqda əslində qradient enmə alqoritmlərinə toxunurduq.ən optimal funksiyalarından sigmoid relu lrelu və s misal çəkə bilərik.

$$MSE=\frac{1}{n}\sum_{i=1}^{n}\left ( \hat{y} - y \right )^2$$

$$\frac{\partial MSE}{\partial \theta}=\frac{1}{n}\sum_{i=1} (y_{i}-\theta^tx_{i})^2 =$$

$$= \frac{2}{n}\sum_{i=1} (y_{i}-\theta^tx_{i})x_{i}    $$

Optimal grad dəyər teta ilə birbaşa əlaqəlidir.teta-nın 0-a doğru yaxınlaşması,MSE nin optimal dəyərini göstərəcəkdir.Və bu da bizə gradient optimal dəyəri verəcək

## Types of logistic regression-Kategoriyaların sayına görə Logistik reqressiya növləri:

1) <strong>binomial</strong>: hədəf dəyişən yalnız iki mümkün növə sahib ola bilər: "qalib" və "zərər", "keçmək" və "uğursuz", "ölü" və "diri" və s. təmsil edə bilən "0" və ya "1" (binar klassifikasiya)

2) <strong>multinomial</strong>: hədəf dəyişən 3 və ya daha çox mümkün növə sahib ola bilər, "xərçəng xəstəliyi" və "xəstəlik B qrupu" və "xəstəlik C qrupu" kimi.

3) <strong>ordinal</strong>-nizamlı: sifariş edilmiş kateqoriyalarla hədəf dəyişənləri ilə birbaşa əlaqəli olması. Məsələn, bir test balı aşağıdakı kimi təsnif edilə bilər: "çox zəif", "yoxsul", "yaxşı", "çox yaxşı". Burada hər kateqoriyaya 0, 1, 2, 3 kimi bir xal verilə bilər.

Qərar necə qəbul edilir!

$$\hat{y} = sign(x\theta)$$

```python
w = np.array([7, -1, 1]) # Ax + By + C = 0
     
# Qeyd: nöqtədən məsafə (x0, y0) -> Ax + By + C  abs(Ax0 + By0 + C) / sqrt(A^2 + B^2)
    
def predict(x, w):
    return np.sign(x.dot(w))
x_pred = np.hstack([np.random.rand(10, 2) * 4 - 2, np.ones(10).reshape(-1, 1)])
y_pred = predict(x_pred, w)

print(x_pred.shape)
print(y_pred.shape)
```

```python
x_naive = np.arange(-0.5, 0.5, 0.1)
# naive boundary
y_naive = 7 * x_naive + 1

plt.scatter(x_pred[:, 0], x_pred[:, 1], c=['orange' if y_i > 0 else 'blue' for y_i in y_pred], marker='x', s=100)
plt.title("Classification")
plt.xlabel("x1 (feature)")
plt.ylabel("x2 (feature)")
plt.scatter(X[:, 0], X[:, 1], c=colors)
plt.plot(x_naive, y_naive, c='red', label='descision boundary naive')
plt.legend();
```

Nəticənin qiymətləndirilməsi

margin - sərhəddə yaxın elementləri dəyərləndirmək bir o qədər çətin olacaq.Elementlər sərhəddən nəqədər uzaq olarsa bir oqədər iki sinif üçün elementləri ayırd edə biləcəyik.Margin,elementin sərhəddə olan məsafəsidir.

$$M_i = y_i \cdot x_i\theta  $$ ( 𝑦=1 və ya 𝑦=−1)

$$M_i \leq 0 \Rightarrow y_i \neq \hat{y_i}$$
$$M_i > 0 \Rightarrow y_i = \hat{y_i}$$

Funksiyanın zərəri(zero-one-loss)

$$f(x) = \begin{cases}
      1, & \text{əgər}\space \hat{y} \neq y, \\
      0, & \text{əgər}\space \hat{y} = y
    \end{cases}$$

Empirik risk:

$$Q(\theta, x) = \frac{1}{n} \sum_{i=1}^{n} f(x) = \frac{1}{n} \sum_{i=1}^{n}[M_i < 0]$$

Problemi yenidən tənzimləyirik

Bir sinif yerinə, bir sinfə aid olma ehtimalını təxmin edirik

$$\hat{p} = \sigma(x\theta)$$

$$\sigma(t)=\frac{1}{1 + exp(-t)}$$

$\sigma - Sigma$

P(Pozitive) sinfinə aid olma ehtimalı olsun

$$OR = \frac{P} {1 - P}$$

$$log {OR} \in R$$

Xətti reqressiya istifadə edərək 𝑙𝑜𝑔𝑂𝑅 üçün bir proqnoz verməyi öyrəndi

$$log OR = x \theta$$

$$P(y_i = 1 | x, \theta) = \frac {OR} {1 + OR} = \frac{e^{x\theta}} {1 + e^{x\theta}} = \sigma(x\theta)$$

$$P(y_i = -1 | x, \theta) = ... = \sigma(-x\theta)$$

$$P(y | x, \theta) = \prod_i {P(y = y_i | x_i, \theta)}$$

```python
t = np.linspace(-10, 10, 100)
sigmoid = 1 / (1 + np.exp(-t))
plt.figure(figsize=(9, 3))
plt.plot([-10, 10], [0, 0], "k-")
plt.plot([-10, 10], [0.5, 0.5], "k:")
plt.plot([-10, 10], [1, 1], "k:")
plt.plot([0, 0], [-1.1, 1.1], "k-")
plt.plot(t, sigmoid, "b-", linewidth=2, label=r"$\sigma(t) = \frac{1}{1 + e^{-t}}$")
plt.xlabel("t")
plt.legend(loc="upper left", fontsize=20)
plt.axis([-10, 10, -0.1, 1.1])
plt.show()
```

```python
xmin, xmax = -4, 4
xx = np.linspace(xmin, xmax, 100)
lw = 2
plt.plot([xmin, 0, 0, xmax], [1, 1, 0, 0], lw=lw, label="Zero-one loss")
plt.plot(xx, np.log2(1 + np.exp(-xx)), color='green', lw=lw, label="Log loss")
plt.xlabel(r"Decision function $f(x)$")
plt.legend(loc="upper right")
plt.ylabel("$L(y=1, f(x))$")
plt.ylim((-1, 6));
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression(solver='lbfgs')
log_reg.fit(X, y)
x0, x1 = np.meshgrid(
        np.linspace(-2, 2, 500).reshape(-1, 1),
        np.linspace(-3, 3, 200).reshape(-1, 1),
    )
X_new = np.c_[x0.ravel(), x1.ravel()]
y_proba = log_reg.predict_proba(X_new)
plt.figure(figsize=(10, 4))
plt.plot(X[y==0, 0], X[y==0, 1], "bs")
plt.plot(X[y==1, 0], X[y==1, 1], "g^")
zz = y_proba[:, 1].reshape(x0.shape)
contour = plt.contour(x0, x1, zz, cmap=plt.cm.brg)
left_right = np.array([-3, 3])
boundary = -(log_reg.coef_[0][0] * left_right + log_reg.intercept_[0]) / log_reg.coef_[0][1]
plt.clabel(contour, inline=1, fontsize=12)
plt.plot(left_right, boundary, "k--", linewidth=3)
plt.xlabel("x1 (feature)")
plt.ylabel("x2 (feature)")
plt.axis([-3, 3, -2, 2])
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
def quality_report(prediction, actual):
    print("Accuracy: {:.3f}\nPrecision: {:.3f}\nRecall: {:.3f}\nf1_score: {:.3f}".format(
        accuracy_score(prediction, actual),
        precision_score(prediction, actual),
        recall_score(prediction, actual),
        f1_score(prediction, actual)
    ))
```

```python
data = pd.read_csv('dataset/column_2C_weka.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
data['class'] = data['class'].apply(lambda x: 1 if x=='Abnormal' else 0)
```

```python
# train-test split
X_train, X_test, y_train, y_test = train_test_split(
    data.drop(['class'], axis=1), data['class'], test_size=0.20, random_state=42, stratify=data['class']
)
```

```python
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# knn
knn = KNeighborsClassifier(n_neighbors=40)
knn.fit(X_train_scaled, y_train)

print("Train quality:")
quality_report(knn.predict(X_train_scaled), y_train)
print("\nTest quality:")
quality_report(knn.predict(X_test_scaled), y_test)
```

Yuxarıda göstərilmiş dataseti knn metodu ilə klassifikasiya etmişdik.Təkrar edərək nəticələrə baxaq.Və sonra logistik reqressiya ilə klassifikasiyanın nəticələrini müqayisə edək

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression()
log_reg.fit(X_train_scaled, y_train)
print("Train quality:")
quality_report(log_reg.predict(X_train_scaled), y_train)
print("\nTest quality:")
quality_report(log_reg.predict(X_test_scaled), y_test)
```

logistik reqressiya vasitəsilə əldə etdiyimiz nəticələr,knn -metoduna nisbətən daha yüksəkdir

```python
featureImportance = pd.DataFrame({"feature": data.drop('class',axis=1).columns, 
                                  "importance": log_reg.coef_[0]})

featureImportance.set_index('feature', inplace=True)
featureImportance.sort_values(["importance"], ascending=False, inplace=True)
featureImportance["importance"].plot('bar', figsize=(9, 7));
```

$$x = \begin{bmatrix} 1 & x_{11} & ... & x_{p1} \\ 1 & x_{12} & ... & x_{p2} \\ ... & ... & ... & ... \\ 1 & x_{n1} & ... & x_{n1}
\end{bmatrix} \quad 
\Theta = \begin{bmatrix} \theta_{01} &  ... & \theta_{0k} \\ \theta_{11} & ... & \theta_{1k} \\ ... & ... & ... \\ \theta_{p1} & ... & \theta_{pk} \end{bmatrix} \quad

$$\hat{p}_k = \frac{e^{x\theta_k}}{\sum_{j=1}^K e^{x\theta_j}}$$

Cross Entropy

$$Q(\Theta, x) = - \frac{1}{n}\sum_{i=1}^{n}\sum_{k=1}^{K} y_{ik} \log \hat{p}_{ik}$$

### Gradient Descent

https://towardsdatascience.com/implement-gradient-descent-in-python-9b93ed7108d1

$$\theta := \theta - \alpha\frac{\partial L}{\partial \theta}$$

* $\alpha$ -  enmənin sürəti

Zərər funksiyasının gradientinin 𝑅𝑆𝑆 (𝜃) nə ilə bərabər olduğunu hesablayırıq

$$RSS = \mathcal{L}(\theta) = (\hat{y} - y)^2$$

$$\frac{\partial L}{\partial \theta_i} = 2(\hat{y} - y)\frac{\partial L}{\partial \theta_i}(\hat{y} - y) = 2(\hat{y} - y)\frac{\partial L}{\partial \theta_i}(\theta_0x_0 + ... + \theta_1x_1 - y) = 2(\hat{y} - y)\cdot x_i$$

$$\theta_i:= \theta_i - \alpha(\hat{y} - y)\cdot x_i$$

$$\frac{\partial RSS(\theta)}{\partial \theta_i} = 2\sum_{i=1}^{n}(\theta^T\cdot x_i - y_i)x_i$$

$$\nabla_\theta RSS(\theta) = \left( \begin{matrix} \frac{\partial L}{\partial \theta_0} \\ \frac{\partial L}{\partial \theta_1} \\ ... \\ \frac{\partial L}{\partial \theta_p} \end{matrix} \right) = x^\top(x\theta - y)$$

MSE - üçün

$$\frac{\partial L}{\partial \theta} = \frac{1}{n} X^\top(X\theta - y)$$

### learning rate - öyrənmə dərəcəsi

```python
class LogisticRegression:
    def __init__(self, lr=0.01, num_iter=100000, fit_intercept=True, verbose=False):
        self.lr = lr
        self.num_iter = num_iter
        self.fit_intercept = fit_intercept
        self.verbose = verbose
    
    def __add_intercept(self, X):
        intercept = np.ones((X.shape[0], 1))
        return np.concatenate((intercept, X), axis=1)
    
    def sigmoid(self, z):
        return 1 / (1 + np.exp(-z))
    def loss(self, h, y):
        return (-y * np.log(h) - (1 - y) * np.log(1 - h)).mean()
    
    def fit(self, X, y):
        if self.fit_intercept:
            X = self.__add_intercept(X)
        
        # weights initialization
        self.theta = np.zeros(X.shape[1])
        
        for i in range(self.num_iter):
            z = np.dot(X, self.theta)
            h = self.sigmoid(z)
            gradient = np.dot(X.T, (h - y)) / y.size
            self.theta -= self.lr * gradient
            
            z = np.dot(X, self.theta)
            h = self.sigmoid(z)
            loss = self.loss(h, y)
                
            if(self.verbose ==True and i % 10000 == 0):
                print(f'loss: {loss} \t')
    
    def predict_prob(self, X):
        if self.fit_intercept:
            X = self.__add_intercept(X)
    
        return self.sigmoid(np.dot(X, self.theta))
    
    def predict(self, X):
        return self.predict_prob(X).round()
```
