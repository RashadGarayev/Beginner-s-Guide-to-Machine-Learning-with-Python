```python
from scipy.spatial.distance import euclidean
#evklid əvəzinə minkovski istifadə edərək nəticəyə baxa bilərsiniz
#from scipy.spatial.distance import minkowski
```

```python
def FindNearestP(new_data, X):
    Dmin = euclidean(new_data, X[0])
    minID = 0
    for i in range(1, X.shape[0]):
        D = euclidean(new_data, X[i])
        if D < Dmin:
            Dmin = D
            minID = i
    return minID


minID = FindNearestP(new_data,X)
```

```python
def PlotPoints(X,y,new_data=None,new_prediction=None, nearest_points=None):
    plt.figure(figsize=(8, 8))
    plt.scatter(X[:, 0], X[:, 1], c=color, s=80,alpha=0.4)
    if new_data is not None:
        plt.scatter(new_data[:,0],new_data[:,1],c='black', s=100)
    if nearest_points is not None:
        plt.scatter(nearest_points[:,0],nearest_points[:,1],c='blue', s=100,edgecolor='black')
    if new_prediction is not None:
        plt.scatter(new_data[:, 0], new_data[:, 1], c='yellow', s=100, edgecolor='black')
    plt.title("Classification 'KNN'")
    plt.xlabel("x1")
    plt.ylabel("x2")
```

```python
from scipy.spatial.distance import cdist
from collections import Counter
def PredictClass(new_data, X, y, k=1):

    dist = cdist(new_data, X)
    N = y[np.argpartition(dist, k, axis=1)[:, :k]]
    return np.array([Counter(row).most_common(1)[0][0] for row in N])
```

```python
y_pred = PredictClass(new_data, X, y, k=3)
```

Bir modelin Doğruluq faizi bizi qane etməsi o modelin ideal olmasına heç də zəmin yaratmır.Çünki accuracy -dən savayı digər parametrlərin də rolu böyükdür.

recall (Həssaslıq) və precision (Dəqiqlik) TP, FP, TN, FN parametrləri önəmlidir.Bu parametrlər information extraction adlanır.

1. TP (True positive — Doğru Pozitiv)
2. FP (False positive — Yanlış Pozitiv)
3. TN (True negative — Doğru Negativ)
4. FN (False negative — Yanlış Negativ)

<strong>1. Accuracy</strong>

Əsas problemi qeyri-bərabər siniflərlə bağlı problemlərdə faydasız olmasıdır. Məsələn, 50 xəstə və 950 sağlam şəxs var. Onları ayırd etməyi öyrənmək istəyirik. Alqoritmimiz hər kəsin sağlam olduğunu təxmin etsin. Bu vəziyyətdə düzgün cavabların nisbəti 95% olacaqdır

<strong>2. Dəqiqlik (Precision)</strong>

P = TP/TP+FP

Alqoritmə nə qədər etibar edə biləcəyimizi əks etdirir

<strong>3. Həssaslıq (recall)</strong>

R = TP/TP+FN

Alqoritm, müəyyən bir sinfi ümumiyyətlə aşkar etmək qabiliyyətini nümayiş etdirir və dəqiqlik bu sinfin digər siniflərdən fərqləndirmə qabiliyyəti dəyəri ilə davam etdirilir.

<strong>4. F-measure və ya F1-score </strong>

F = 1/B*(1/p)+(1-B)*1/R

Bu vəziyyətdə 𝛽 parametr, metrikdə dəqiqliyin ağırlığını təyin edir. 𝛽 = 1 eyni çəkilərlə dəqiqliyi və tamlığı təqdim edir. Dəyişdirərək, birinə və ya digərinə üstünlük verə bilərsiniz.

## Model parametrlərinin seçilməsi

<strong>1. Testing set - </strong>Təlimdən öncə(training) məlumatların bir hissəsini təxirə salırıq və model təlim mühitinə təqdim olunana qədər istifadə etmirik. Hazır modelin keyfiyyətini yoxlamaq üçün son və etibarlı yoldur

<strong>2. Training set - </strong>Birbaşa model hazırlamaq üçün istifadə olunur, model bu məlumatları təlim zamanı görür.

<strong>3. Validation set - </strong>Model, təlim zamanı bu məlumatları görmür, lakin biz hiperparametrləri konfiqurasiya etmək üçün istifadə edirik, yəni təlim keçmiş modelin keyfiyyətini yoxlayırıq və bundan narazıyıqsa, model parametrlərini dəyişdiririk və proseduru təkrarlayırıq. Beləliklə, model bu məlumatlarla birbaşa işləmir, yenə də onlardan gələn məlumatları eyni şəkildə istifadə edirik və konfiqurasiya edilmiş parametrlər şəklində modelə sızır.

Yeni dataset hazırlayaq

```python
X, y = datasets.make_blobs(n_samples=100, random_state=5, centers=2)
PlotPoints(X, y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)#datasetdə olan elementlərin yüzdə qırx faizini(test_size=0.4) test seti üçün istifadə edirik
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape 
```

Xatırlayırsınızsa k-parametrini funksiyaya tək-tək yazaraq nəticəni əldə edirdik.Bunun əvəzinə dövr operatoru tərtib edib,müəyyən aralıqda ədədləri k-ya verərək təlimin hansı parametrdə daha yaxşı model hazırladığını öyrənə bilərik

```python
neighbors = range(1, 50)
accuracy_train = []
accuracy_test = []
for k in neighbors:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    accuracy_train.append(accuracy_score(knn.predict(X_train), y_train))
    accuracy_test.append(accuracy_score(knn.predict(X_test), y_test))
    
plt.plot(neighbors, accuracy_train, color='red', label='train')
plt.plot(neighbors, accuracy_test, color='green', label='test')
plt.title("Max test quality: {:.3f}\nBest K: {}".format(max(accuracy_test), np.argmax(accuracy_test)+1))
plt.legend()
plt.show()
```

Yuxarıdakı qrafikdə modelin test və training ərəfəsində gurultulu olduğunun şahidi oluruq.Və k-5 dəyərinin dataset üçün ən yaxşı dəyər oluğunu öyrəndik.Doğruluq isə 0.75 göstərilmişdir

## K-qat çarpaz təsdiqləmə (k-fold cross-validation) validasiya medotu

Əvvəlki bəhsdə dataseti təlim və test seti olaraq ikiyə ayırmışdıq.Modeli qiymətləndirmə üçün test setindən istifadə edirdik.Misal olaraq təlim sətinə 66% test setində isə 33% məlumatların olması.Bu metoddan başqa validasiya,çarpaz təsdiqləmə metodu da geniş istifadə olunur

çarpaz təsdiqləmə metodunda öncə K-üçün dəyər seçilir,misal olaraq k-3 seçərsək veriləni 3sayda bərabər hissələrə ayırırıq.

```python
from IPython.display import Image
Image(filename='image/Untitled 2.png',width=400,height=400)
```

Yuxarıda göstərilən qaydada,k-dəyəri qədər hissələrə ayırırıq.Və parçalara ayırdıqdan sonra,k-qat çarpaz təsdiqləmə çalışmağa başlayacaqdır.Bu hissələrdən biri test üçün digər iki hissə isə təlim üçün nəzərdə tutulacaq.Bu hissələrin hansının seçilməsinin vacibliyi yoxdur.Misal olaraq ilk iki hissəni təlim üçün üçüncü hissəni isə test üçün seçə bilərik.

Və hissələrə böldükdən sonra təlimə start veririk.İlk hissə təlim üçün nəzərdə tutulduğundan nəticəni alıb növbəti hissə üçün təlim başlayacaqdır.Nəticə etibarı ilə 3 təlim və test setini çalışdırdıqdan sonra əldə etdiyimiz doğruluq və ya xəta payı(error rate) bu üç nəticənin ortaq qiymətinin alınması ilə ölçülür.

```python
#Ehtiyacımız olan kitabxananı import edirik
from sklearn.model_selection import cross_val_score
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
scores = cross_val_score(knn, X, y, cv=3, scoring='accuracy')
print(scores)
print('Mean', scores.mean())
print('Std', scores.std())
```

Yuxarıdakı nümunədə cv-3 qeyd edərək 3 hissəyə bölünməsini təmin etdik.Qeyd edimki hissələrə bölmə prosesini k-qiyməti ilə eyni olmasının vacibliyi yoxdur.scoring -parametrinə isə doğruluq haqqında məlumatı almaq istədiyimizi yazdıq.Və ya beş hissəyə ayırıb nəticəyə baxaq

```python
knn = KNeighborsClassifier(n_neighbors=3)
scores = cross_val_score(knn, X, y, cv=5, scoring='accuracy')
print(scores)
print('Mean:', scores.mean())
print('Std:', scores.std())
```

k-fold  5 hissəyə ayırdıqda isə acc-doğruluq qiymətləri cv-3 -ə görə daha yüksəkdir.

## GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
#n_neighbors or k range(1,50)
param_grid = {
        'n_neighbors': range(1, 50)}

knn = KNeighborsClassifier()
grid_search = GridSearchCV(knn, param_grid, cv=5, scoring='f1')
grid_search.fit(X_train_scaled, y_train)
print("Best CV score: {:.3f}, best CV k: {}".format(grid_search.best_score_, grid_search.best_estimator_.n_neighbors)) 
test_predictions = grid_search.best_estimator_.predict(X_test_scaled)
print("Resulting test score: {:.3f}".format(f1_score(test_predictions, y_test)))
```

Ayrı-ayrı funksiya vasitəsilə həm k parametrini eləcədə k-fold hissələrə bölmə istifadə etmək yerinə GridSearchCV -metodunu istifadə edərək hiperparametrləri daxil edə bilərik

```python
from sklearn.model_selection import RandomizedSearchCV

param_grid = {
    'n_neighbors': range(1, 50)}

knn = KNeighborsClassifier()

rand_search = RandomizedSearchCV(knn, param_grid, cv=5, scoring='f1', random_state=40)
rand_search.fit(X_train_scaled, y_train)

print("Best CV score: {:.3f}, best CV k: {}".format(
    rand_search.best_score_, rand_search.best_estimator_.n_neighbors)
) 

test_predictions = rand_search.best_estimator_.predict(X_test_scaled)
print("Resulting test score: {:.3f}".format(f1_score(test_predictions, y_test)))
```
