# SVM (support vector machine)  - Dəstək vektor maşınları

### İntuisiya

Maşın öyrənmə alqoritmlərini mənimsəmək ümumiyyətlə çətin deyil. Yeni başlayanların əksəriyyəti reqressiyanı öyrənməklə başlayır.Öyrənmək və istifadə etmək çox sadədir, lakin bu bizim məqsədimizi həll etmir.Çünki yalnız Reqressiyadan daha çox şey edə bilərsiniz.Maşın öyrənmə alqoritmlərini balta, qılınc, bıçaq, yay, xəncər və s. kimi müxtəlif vasitələrə bənzədə bilərsiniz, ancaq onları vaxtında istifadə etməyi öyrənməlisiniz.Bir bənzətmə olaraq, 'Reqressiya' məlumatları səmərəli şəkildə kəsib dəyərləndirməyə qadir olan, lakin olduqca mürəkkəb məlumatlarla işləməyi bacarmayan bir qılınc kimi düşünün. Əksinə, 'Dəstək Vektor Maşınları' kəskin bıçaq kimidir - daha kiçik məlumat bazaları üzərində işləyir, lakin mürəkkəb olanlarda maşın öyrənmə modellərini qurmaqda çətinlik çəkə bilər.Hər iki halda çeşid çoxdur.Əlimizdə alqoritmlər var.uyğun yerə uyğun metodu istifadə edərək möhtəşəm nəticələr əldə edə bilərik.

Dəstək vektor Maşınları metodu 1960-cı illərdə aid olsada 95-ci ildən Vladir Vapnik, Berhard Boser və Isabelle Guyon tərəfindən inkişaf etdirilmişdir.Dərin Öyrənmə dərslərində siz bu vektor maşınlarından geniş istifadə edəcəksiniz.Misal olaraq Rəsmlərin sinifləndirilməsində (İmage classification) seçilmiş region və neqativ bölgələr ayrılmaqla svm-dən keçirilir və nəticədə predict əldə olunur.

"Dəstək Vektor Maşınları" (SVM) həm təsnifat, həm də reqressiya problemləri üçün istifadə edilə bilən nəzarət edilən bir maşın öyrənmə alqoritmidir. Bununla birlikdə, əsasən təsnifat problemlərində istifadə olunur.
İki sinifi çox yaxşı fərqləndirən hiper müstəvini taparaq təsnifat aparırıq.
Qısa desək Dəstək vektor maşınları (SVM) təsnifat(sinifləndirmə), reqressiya və xarici satışların aşkarlanması üçün istifadə olunan nəzarət olunan(Rəhbərli Öyrənmə) təlim metodları toplusudur. Dəstək vektor maşınlarının üstünlükləri aşağıdakılardır:

<p>1) Yüksək ölçülü fəzalarda təsirli.</p>
<p>2) Ölçülərin sayı nümunələrin sayından çox olduğu hallarda təsirli olur.</p>
<p>3) Qərar funksiyasında  yaddaş səmərəliyi mövcuddur.
<p>4) Qərar funksiyası üçün fərqli Kernel funksiyaları təyin edilə bilər.</p>

Dəstək vektor maşınları xətti və qeyri xətti(Linear and Nonlinear) olmaqla sinifləndirilmə metodunu tətbiq oluna bilinər.

İkili sinifləndirmədə(binary classification) əsasən tərəflərdən biri(datasetdə olan vektorlar) 1,digəri -1 olmaqla sinifləndirilir.Başqa cür ifadə etsək pozitiv və neqativ.Sinifləndirmə metodunu tətbiq etmək üçün iki sinif arasında birbirinə yaxın iki əsas dəstək vektoru seçilir,və bu d-vektorlardan margin tətbiq etməklə hər iki vektor sərhədinə bərabər-paralel şəkildə əsas sərhəd(y=0) çəkilir.tərəflərdən biri yəni d-vektorundan keçən sərhəd y=1 digəri isə y=-1 qaydada ifadə edilir

Loqistik reqressiya dərslərini xatırlayırsınızsa,iki sinif arasında optimal decision boundary -sərhəd tərtib edilirdi.Dəstək vektor maşınlarında isə öncə dəstək vektorları üçün sərhəd seçilir,daha sonra optimal sərhəd tərtib olunur.

Hər zaman iki vektor bir-birindən son dərəcə ideal formada ayrılmaya bilir.təyin etdiyimiz ən optimal hiper doğruluq xətdi,üst və aşağı hissədə hər iki sinifdən məlumatlar üçün ideal olmaya da bilir.Məsələn +1 kimi qeyd etdiyimiz class A sinfində class B elementləri də ola bilər.Biz bu məqamı tam ayrılmayan məlumatlar toplusu adlandırırıq.

Məlumatlarımızı iki sinif olmaq üzrə(class A class B) hiper-doğruluq xətdini fərz edəkki qeyd etdik

class A üçün , $w^{T}x_i + w_0 \geq 1$ şərti üçün

$y_i((w^{T}x_i) + w_0) \geq 1 - \xi_i$

kimi ifadə etdikdə ksi dəyəri $\xi_i=0$ olacaq

class C ifadəsi üçüncü hər hansı sinif mənasını ifadə etmir.vektor margin-məsafə payının tam aralığında qalır.Yuxarıda qeyd etdiyimiz kimi hər zaman ideal datasetlərinə rast gəlməyəcəyik.Bu zaman C üçün düsturu yazsaq

$y_i((w^{T}x_i) + w_0) \geq 1 - \xi_i$ 
<p>$0 < \xi_i <1$

Bu halda ksi dəyəri sıfır və 1 aralığında qiymət ala bilər

class B üçün , $w^{T}x_i + w_0 \leq 0$ şərti üçün

$y_i((w^{T}x_i) + w_0) \geq 1 - \xi_i$

Bu halda isə ksi dəyəri 1ə bərabər və ya birdən böyük qiymətlər alacaqdır

$\xi_i \geq 1$

və gördüyünüz kimi $\xi_i$ qiymət verərək məsafə payını dəyişdirə bilirik.Vacib məsələ ondan ibarətdirki genişləmə payını kiçiltmək və böyütməklə birbaşa optimallaşdırma əməliyyatına təsir edə bilərik.Daha böyük götürsək hər iki siniflərə aid məlumatları qəbul edə bilər və ya daha kiçik götürsək vacib olan elementlər kənarda qala bilər.

Optimizasiya

<strong>Ən böyük 

$ \frac{2}{||\omega||}$

$y_i((w^{T}x_i) + w_0) \geq 1 $ 
$i=1,......n$

<strong>Ən kiçik

$ \frac{1}{2}{||\omega||^{2}}$ 
və ya 
$\frac{1}{2}{\omega^{T}\omega}$

$y_i((w^{T}x_i) + w_0) \geq 0 $ 

Düsturumuzda artıq dəyişiklik etməliyik.Yuxarıda ksi ifadəsindən bəhs etmişdik.Və düstura ksi ifadəsini də əlavə etməliyik.Bu zaman

$y_i((w^{T}x_i) + w_0) \geq 1 - \xi_i$  
$i = 1,........n$

$ \frac{1}{2} ||w||^2 + C\sum\limits_i\xi_i^{d} $

$\xi_i \geq 0$
$i=1,......n$

Düsturda ksi($\xi_i$) üzəri d - ifadəsi 1 qiyməti,bəzəndə kvadratı şəklində ifadə olunur.C - parametri isə təyin edəcəyimiz soft margin (hard margin) dəyəridir.C-parametrini təqribi uyğun seçmək önəmlidir.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets.samples_generator import make_classification

from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from scipy import stats
import pandas as pd
# use seaborn plotting defaults
import seaborn as sns; sns.set()
plt.rcParams["figure.figsize"] = (12, 8)
pd.options.display.float_format = '{:.2f}'.format
```

Buna nümunə olaraq, iki sinif nöqtəsinin yaxşı ayrıldığı bir təsnifat tapşırığının sadə vəziyyətini nəzərdən keçirək:

```python
from sklearn.datasets.samples_generator import make_blobs
X, y = make_blobs(n_samples=50, centers=2,
                  random_state=0, cluster_std=0.60)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn');
```

Xətti  təsnifatçısı iki məlumat dəstini ayıran düz bir xətt çəkməyə çalışır və beləliklə təsnifat üçün bir model yaradır.Burada problem görürük: iki sinif arasında mükəmməl ayırd edə biləcək birdən çox mümkün bölüşdürmə xətti var.

```python
xfit = np.linspace(-1, 3.5)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn')
plt.plot([0.6], [2.1], 'x', color='green', markeredgewidth=2, markersize=10)

for m, b in [(1, 0.65), (0.5, 1.6), (-0.2, 2.9)]:
    plt.plot(xfit, m * xfit + b, '-b')

plt.xlim(-1, 3.5);
```

x - ilə ifadə olunan xətdimiz daha ideal ayırdedici görünür.Marjı maksimum dərəcədə artırmaqla dəstək vektorlarını təyin etməimkanına malik ola bilərik

```python
xfit = np.linspace(-1, 3.5)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='autumn')

for m, b, d in [(1, 0.65, 0.33), (0.5, 1.6, 0.55), (-0.2, 2.9, 0.2)]:
    yfit = m * xfit + b
    plt.plot(xfit, yfit, '-k')
    plt.fill_between(xfit, yfit - d, yfit + d, edgecolor='none',
                     color='#AAAAAA', alpha=0.3)

plt.xlim(-1, 3.5);
```

və ya dəstək vektorlarımızı nəzərə alaraq

```python
def plot_svc_log_decision_function(clf1, clf2, ax=None):
    if ax is None:
        ax = plt.gca()
    x = np.linspace(plt.xlim()[0], plt.xlim()[1], 30)
    y = np.linspace(plt.ylim()[0], plt.ylim()[1], 30)
    XX, YY = np.meshgrid(x, y)
    XY = np.c_[XX.ravel(), YY.ravel()]
    P1 = clf1.decision_function(XY)
    P1 = P1.reshape(XX.shape)
    
    P2 = clf2.decision_function(XY)
    P2 = P2.reshape(XX.shape)
    cplot = ax.contour(XX, YY, P1, colors='k', label='svm',
               levels=[-1, 0, 1], alpha=0.5,
               linestyles=['--', '-', '--'])
    ax.clabel(cplot, inline=1, fontsize=10)
    
    ax.contour(XX, YY, P2, colors='r', label='logreg',
               levels=[0], alpha=0.5,
               linestyles=['-'])

    
def plot_svc_decision_function(clf1, ax=None):
    if ax is None:
        ax = plt.gca()
    x = np.linspace(plt.xlim()[0], plt.xlim()[1], 30)
    y = np.linspace(plt.ylim()[0], plt.ylim()[1], 30)
    XX, YY = np.meshgrid(x, y)
    XY = np.c_[XX.ravel(), YY.ravel()]
    P1 = clf1.decision_function(XY)
    P1 = P1.reshape(XX.shape)
    
    cplot = ax.contour(XX, YY, P1, colors='k', label='svm',
               levels=[-1, 0, 1], alpha=0.5,
               linestyles=['--', '-', '--'])
    ax.clabel(cplot, inline=1, fontsize=10)
    

def lin_sep_svm_demo(class_sep=2):
    X, y = make_classification(n_samples=100, n_features=2, n_informative=2, class_sep=class_sep, scale=1,
                                n_redundant=0, n_clusters_per_class=1, random_state=31)

    lin_svm = SVC(kernel='linear', C=100).fit(X, y)
    
    log_reg = LogisticRegression(C=100).fit(X, y)
    
    plt.figure(figsize=(10,7))
    plt.scatter(X[:, 0], X[:, 1], c=y, s=70, cmap='autumn')
    plot_svc_log_decision_function(lin_svm, log_reg)

    plt.xlabel('$x_1$')
    plt.ylabel('$x_2$')
    
    plt.xlim(-2, 5)
    plt.ylim(-3, 4)
```
